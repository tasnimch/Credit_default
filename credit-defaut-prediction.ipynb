```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_excel('../input/default/default.xlsx')
df=df1.copy()
```

```python
pd.to_datetime(df['DATE_DE_CREATION_TIERS'],format='%D%M%Y')
pd.to_datetime(df['DATE_DE_CREATION_ENTREP'],format='%D%M%Y')
df.drop('numtiers',axis=1,inplace=True)
df.drop('Annee',axis=1,inplace=True)
df.set_index('NUMTIERS_ANNEE')
```

```python
df['DIVERSITE_CLIENTS'].replace("Diversification tres forte par produits, clients, situation geographique",'Diversification trés forte par produits, clients, situation geographique',inplace=True)
df['DIVERSITE_FOURNISSEURS'].replace('Tres grande diversite','Très grande diversite',inplace=True)
df['DIVERSITE_FOURNISSEURS'].replace('Trés grande diversite','Très grande diversite',inplace=True)
df['REPUTATION'].replace('Tres bonne','Très bonne',inplace=True)
df['NIVEAU_COMPETITIVITE'].replace('Tres forte concurrence','Très forte concurrence',inplace=True)

df['IMPACT_SOCIAUX_ENVIRONNEMENTAL'].replace('Aucun impact social ou environnemental, soumis e une reglementation','Aucun impact social ou environnemental, soumis à une reglementation',inplace=True)
    
    
```

```python
categories_features=[]
numeric_features=[]
time_series_features=[]
for feature in df.columns:
    if df[feature].dtype=='O':
        categories_features.append(feature)
    elif  (df[feature].dtype in('int64',"float64")) and (feature!='defaut'):
        numeric_features.append(feature)
    else:
        time_series_features.append(feature)
print(len(categories_features),'///',len(numeric_features),'///',len(time_series_features))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le_count = 0
for col in categories_features:
    if len(set(df[col])) <= 4:
        le.fit(df[col].dropna())
        df[col] = le.transform(df[col])
        le_count += 1
            
print('%d columns were label encoded.' % le_count)
```

```python
target=df['defaut']
df.drop('defaut',inplace=True,axis=1)
```

```python
import featuretools as ft
es=ft.EntitySet(id='default')
es = es.add_dataframe(
    dataframe_name="default",
    dataframe=df,
        index="NUMTIERS_ANNEE",

)

es
```

```python
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="default",
    agg_primitives=['entropy',"count",'mean','mode'],
    trans_primitives=['divide_numeric'],
    max_depth=5,
)
feature_matrix
```

```python
feature_matrix=pd.get_dummies(feature_matrix)
```

```python
for feature in feature_matrix.columns:
    if (feature_matrix[feature].isna().sum()/1521)>0.2:
        feature_matrix.drop(feature,axis=1,inplace=True)
feature_matrix.replace(np.inf,0.00001,inplace=True)
feature_matrix.replace(-np.inf,0.00001,inplace=True)
feature_matrix.replace(np.nan,0.00001,inplace=True)
feature_matrix.replace(0,0.00001,inplace=True)
```

```python
import pandas as pd
import numpy as np

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame)
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
feature_matrix=clean_dataset(feature_matrix)
```

```python
feature_matrix.select_dtypes('O').columns
```

```python
feature_matrix.index=feature_matrix.index.astype('int')
feature_matrix=feature_matrix.astype('float64')
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer()
imputer.fit_transform(feature_matrix)
```

```python
from sklearn.decomposition import PCA
transformer = PCA(n_components=10)
X_transformed = transformer.fit_transform(feature_matrix.values)
np.sort(transformer.explained_variance_ratio_)
```

```python
X=pd.DataFrame(X_transformed)
```

```python
X['defaut']=list(target)
```

```python
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
# setting up testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,0:2], X['defaut'], test_size=0.25, random_state=27)

# concatenate our training data back together
df_train = pd.concat([X_train, y_train], axis=1)

# Separate majority and minority classes
df_majority = df_train[df_train.defaut==0]
df_minority = df_train[df_train.defaut==1]

# oversample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,    
                                 n_samples=len(df_majority),   
                                 random_state=1234) 
 

df_upsampled = pd.concat([df_majority, df_minority_upsampled])

df_upsampled.defaut.value_counts()
```

```python
X_train=df_upsampled.drop('defaut',axis=1)
y_train=df_upsampled['defaut']
```

```python
len(y_test[y_test==0])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis,LinearDiscriminantAnalysis
from sklearn.metrics import roc_auc_score,f1_score,confusion_matrix,classification_report
RF=RandomForestClassifier()
QDE=QuadraticDiscriminantAnalysis()
LDA=LinearDiscriminantAnalysis()
RF.fit(X_train,y_train)
y_pred=RF.predict(X_test)
f1=f1_score(y_pred,y_test,average='weighted')
print('%.5f' %f1)
print(confusion_matrix(y_pred,y_test))
```
